```python
import pandas as pd
```

```python
df = pd.read_csv('./labeled_gofundme.csv', index_col = 0)
df.head()
```

```python
df.drop(['web-scraper-order', 'web-scraper-start-url', 'link click', 'link click-href'], axis=1, inplace=True)
df.head()
```

```python
def currentSplit(x):
    line = x.split('raised of')
    return line[0].strip().strip('$').replace(',', '')

df['current_raised'] = df['current'].apply(currentSplit)
df.head()
```

```python
def goalSplit(x):
    line = x.split('raised of')
    new_line = line[1].split(' ')
    return new_line[1].strip('$').replace(',', '')

df['goal_fund'] = df['current'].apply(goalSplit)
df.head()
```

```python
df.drop('goal', axis=1, inplace=True)
df.head()
```

```python
df.drop(['current', 'image-src'], axis = 1, inplace=True)
df.head()
```

```python
df.rename(columns = {'intial-po' : 'post'}, inplace=True)
df.head()
```

```python
def isSelf(x):
    import re
    if re.findall('behalf of', x):
        return 0
    else:
        return 1

df['Self-raised'] = df['fundraiser'].apply(isSelf)
df.head()
```

```python
df['current_raised'] = df['current_raised'].apply(lambda x: x.replace(',', ''))
df.head()
```

```python
df['goal_precentage'] = round(df['current_raised'].astype('int') / df['goal_fund'].astype('int'),2)
df.head()
```

```python
df.to_csv('cleaned_gofundme.csv')
```
